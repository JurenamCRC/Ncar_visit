```python
geometry_file = '/glade/scratch/bachman/ETP.012_tides3/run/ocean_geometry.nc'
dsg = xr.open_dataset(geometry_file)
```

```python
dsg=dsg.rename({'lonh':'xh','lonq':'xq','lath':'yh','latq':'yq'})
```

```python
j_eq = (np.abs(dsg['yq'])).argmin().values
print(j_eq,dsg['yq'][j_eq].values,dsg['f'][j_eq,0].values)
```

```python
sclfac = 86400.
rho0 = 1.0e3
grav = 9.81

lonbeg = 273
lonend = 278
latbeg = 7
latend = 12
```

```python
dsg['wet'].sel(xh=slice(lonbeg,lonend),yh=slice(latbeg,latend)).plot()
```

```python
path_in = '/glade/campaign/cgd/oce/people/bachman/ETP_1_20_tides/HOURLY/'
date_str = '*2018_1??'
file_list = path_in + date_str + '.nc'
files = sorted(glob.glob(file_list))

ds = xr.open_mfdataset(files,
                       combine='by_coords',
                       compat='override',
                       coords='minimal',
                       parallel=True,
                       chunks={'time':12}
                      )
```

```python
ds = ds.merge(dsg,compat='equals')
```

```python
ds
```

```python
wet = ds['wet'].astype('int')
land = 1 - wet
e = land.shift(xh=-1,fill_value=1)
w = land.shift(xh=1,fill_value=0)
n = land.shift(yh=-1,fill_value=0)
s = land.shift(yh=1,fill_value=0)
ne = n.shift(xh=-1,fill_value=1)
nw = n.shift(xh=1,fill_value=0)
se = s.shift(xh=-1,fill_value=0)
sw = s.shift(xh=1,fill_value=0)
neighbors = n + s + e + w + ne + nw + se + sw
coast = neighbors*wet
coast = xr.where(coast >= 1,1,0)
coast.attrs = {'long_name':'Mask for coastal ocean points','units':'nondim'}
ocean = wet - coast
ocean.attrs = {'long_name':'Mask for open ocean points','units':'nondim'}
coast.sel(xh=slice(lonbeg,lonend),yh=slice(latbeg,latend)).plot()
```

```python
grid = Grid(ds, 
            coords={'X': {'center': 'xh', 'outer': 'xq'},
                    'Y': {'center': 'yh', 'outer': 'yq'},
                    },
            periodic=None
           )
```

```python
f_at_u = grid.interp(ds['f'],'Y',boundary='fill')
f_at_v = grid.interp(ds['f'],'X',boundary='fill')
```

```python
taux = ds['taux'].fillna(0.)
tauy = ds['tauy'].fillna(0.)
```

```python
taux_at_q = grid.interp(taux,'Y',boundary='extrapolate')
tauy_at_q = grid.interp(tauy,'X',boundary='extrapolate')
```

```python
Uek_at_q =  tauy_at_q/(rho0*ds['f'])
Uek_at_q = Uek_at_q.where(np.isfinite(Uek_at_q),0.)
Vek_at_q = -taux_at_q/(rho0*ds['f'])
Vek_at_q = Vek_at_q.where(np.isfinite(Vek_at_q),0.)
```

```python
Uek_at_u = grid.interp(Uek_at_q,'Y',boundary='fill')
Vek_at_v = grid.interp(Vek_at_q,'X',boundary='fill')
```

```python
Uek_at_u = Uek_at_u.fillna(0.)
Vek_at_v = Vek_at_v.fillna(0.)
```

```python
Wek_u = ((grid.diff(Uek_at_u*ds['dyCu'],'X',boundary='fill'))/ds['Ah'])*ds['wet']
Wek_v = ((grid.diff(Vek_at_v*ds['dxCv'],'Y',boundary='fill'))/ds['Ah'])*ds['wet']
Wek = Wek_u + Wek_v
```

```python
fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10),constrained_layout=True)

ctau=ax[0,0].pcolormesh(ds['xq'].sel(xq=slice(lonbeg,lonend)),ds['yq'].sel(yq=slice(latbeg,latend)),
                   taux_at_q.sel(xq=slice(lonbeg,lonend),yq=slice(latbeg,latend)).mean(dim='time'),
                   vmin=-0.15,vmax=0.15,cmap='bwr')
ax[0,0].set_title(r'$\tau^{x}$')

ax[1,0].pcolormesh(ds['xq'].sel(xq=slice(lonbeg,lonend)),ds['yq'].sel(yq=slice(latbeg,latend)),
                   tauy_at_q.sel(xq=slice(lonbeg,lonend),yq=slice(latbeg,latend)).mean(dim='time'),
                   vmin=-0.15,vmax=0.15,cmap='bwr')
ax[1,0].set_title(r'$\tau^{y}$')
cbar = plt.colorbar(ctau,ax=ax[1,0])
cbar.ax.set_ylabel('Pa')

cv=ax[0,1].pcolormesh(ds['xh'].sel(xh=slice(lonbeg,lonend)),ds['yq'].sel(yq=slice(latbeg,latend)),
                   Vek_at_v.sel(xh=slice(lonbeg,lonend),yq=slice(latbeg,latend)).mean(dim='time'),
                   vmin=-4,vmax=4,cmap='bwr')
ax[0,1].set_title(r'$V_{Ek}$')

ax[1,1].pcolormesh(ds['xq'].sel(xq=slice(lonbeg,lonend)),ds['yh'].sel(yh=slice(latbeg,latend)),
                   Uek_at_u.sel(xq=slice(lonbeg,lonend),yh=slice(latbeg,latend)).mean(dim='time'),
                  vmin=-4,vmax=4,cmap='bwr')
ax[1,1].set_title(r'$U_{Ek}$')
cbar = plt.colorbar(cv,ax=ax[1,1])
cbar.ax.set_ylabel(r'$m^{2} s^{-1}$')
```

```python
h_at_q = grid.interp(grid.interp(ds['ssh'],'X',boundary='fill'),'Y',boundary='fill')
```

```python
ug = -(grav/f_at_u)*grid.diff(h_at_q,'Y',boundary='fill')/ds['dyCu']
vg =  (grav/f_at_v)*grid.diff(h_at_q,'X',boundary='fill')/ds['dxCv']
```

```python
uag = ds['ssu'] - ug
vag = ds['ssv'] - vg
```

```python
ekman_depth = 20.0
g
uek = Uek_at_u/ekman_depth # transport -> velocity
ulinear = ug + uek
vek = Vek_at_v/ekman_depth # transport -> velocity
vlinear = vg + vek
```

```python
Wg_u = ((grid.diff(ug*ds['dyCu'],'X',boundary='fill'))/ds['Ah'])*ds['wet']*ekman_depth
Wg_v = ((grid.diff(vg*ds['dxCv'],'Y',boundary='fill'))/ds['Ah'])*ds['wet']*ekman_depth
Wg = Wg_u + Wg_v
```

```python
ds_out = xr.Dataset()
ds_out['wet'] = wet
ds_out['mask_coast'] = coast
ds_out['mask_openocean'] = ocean

Uek_at_u = Uek_at_u.astype('float32')
Uek_at_u.attrs = {'long_name':'Zonal Ekman transport on east face of h cell','units':'m^2/s'}
ds_out['Uek'] = Uek_at_u

Vel_at_v = Vek_at_v.astype('float32')
Vek_at_v.attrs = {'long_name':'Meridional Ekman transport on north face of h cell','units':'m^2/s'}
ds_out['Vek'] = Vek_at_v

Wek = Wek.astype('float32')
Wek.attrs = {'long_name':'Total Ekman upwelling','units':'m/s'}
ds_out['Wek'] = Wek

Wek_u = Wek_u.astype('float32')
Wek_u.attrs = {'long_name':'Ekman upwelling due to zonal divergence','units':'m/s'}
ds_out['Wek_u'] = Wek_u

Wek_v = Wek_v.astype('float32')
Wek_v.attrs = {'long_name':'Ekman upwelling due to meridional divergence','units':'m/s'}
ds_out['Wek_v'] = Wek_v

ds_out
```
