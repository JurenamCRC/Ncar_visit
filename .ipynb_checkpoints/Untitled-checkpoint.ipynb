```python
files1 = '/glade/scratch/pmora/monthly_wind_stress_curl.nc'
```

```python
DS1 = xr.open_mfdataset(
    files1,
    compat="override",
    data_vars="minimal",
    coords="minimal",
    parallel=True,
    concat_dim="time",
    combine="nested",
)
DS1= DS1["wind_stress_curl"]
DS1
```

```python
DSs = DS1.sel(xh=slice(273.5, 274.75), yh=slice(9.5, 11.5))              
DSs
```

```python
DSs.plot.contourf(xh='lon',yh='lat',col='month',col_wrap=3)
```

```python
count = 0
fig=plt.figure(figsize=(11,13))
for mes in DSs.month:
             count+=1
             ax=fig.add_subplot(4,3,count,projection=cr.PlateCarree())
             DSs[DSs.month==mes].plot(vmin=-4, vmax=4, cmap="coolwarm_r", cbar_kwargs={"label": "w [m/day]"}, levels=np.arange(-4,4.5,0.5))
#             annotate_stats(plt.gca(), Ep2[Ep2.month==mes], x=0.3, y=0.4)
             ax.coastlines(resolution='10m')
             ax.add_feature(cfeature.BORDERS)
             ax.set_extent([-86.5, -85.25, 9.5,11.5])
```
