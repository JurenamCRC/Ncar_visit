```python
import netCDF4 as nc
import numpy as np
import matplotlib.pyplot as plt
```

```python
d=nc.Dataset("/glade/scratch/pmora/mean_temp.nc","r")
temp=d['mean_temp'][:,450:550,519:820]
z_i=d['z_i'][:]
d.close()
```

```python
temp=np.array(temp,dtype=np.float64)
dpt=np.array(z_i,dtype=np.float64)
np.shape(temp)
```

```python
fig, ax = plt.subplots()
ax.plot(temp[0:63,30,182], -1*z_i[0:63], linewidth=1.5,color='r')
plt.rc('grid', linestyle="--", color='gray', linewidth=1)
listOf_Yticks = np.arange(0, -220, -20)
listOf_Xticks = np.arange(10, 32, 2)
plt.yticks(listOf_Yticks)
plt.xticks(listOf_Xticks)
plt.tick_params(axis='both', which='major', labelsize=16)
plt.grid(True)
plt.rcParams["figure.figsize"]=(10,10)
plt.title("Temperature CRD (lat~9°N, lon~-90°W)", fontsize=19)
plt.ylabel("depth (m)", fontsize=17)
plt.xlabel("sea water potential tempeature (°C)", fontsize=17)
ax.tick_params(axis="x", top=True, labelbottom=False, labeltop=True)
```

```python
fig, ax = plt.subplots()
ax.plot(temp[0:28,30,182], -1*z_i[0:28], linewidth=1.5,color='r')
plt.rc('grid', linestyle="--", color='gray', linewidth=1)
listOf_Yticks = np.arange(0, -62, -10)
listOf_Xticks = np.arange(14, 32, 2)
plt.yticks(listOf_Yticks)
plt.xticks(listOf_Xticks)
plt.tick_params(axis='both', which='major', labelsize=13)
plt.grid(True)
plt.rcParams["figure.figsize"]=(8,10)
plt.title("Temp profile near center CRD (lat~9°N, lon~-90°W)", fontsize=17)
plt.ylabel("depth (m)", fontsize=17)
plt.xlabel("sea water potential tempeature (°C)", fontsize=17)
ax.tick_params(axis="x", top=True, labelbottom=False, labeltop=True)
```
