```python
files = '/glade/p/nsc/ncgd0048/ETP_1_20/COMPRESSED_OCEAN_HOURLY/ocean_hourly_*.nc'
```

```python
DS = xr.open_mfdataset(
    files,
    compat="override",
    data_vars="minimal",
    coords="minimal",
    parallel=True,
    concat_dim="time",
    combine="nested",
    chunks={  "yh":200, "xh":200, },
)[["ssh", "sst", "sss","ssu", "ssv" ]]
#[["so", "thetao", "uo", "vo", "wo"]]
```

```python
DS
```

```python
lonbeg = 273
lonend = 278
latbeg = 7
latend = 12
```

```python

month= DS.sel(xq=slice(lonbeg,lonend), xh=slice(lonbeg,lonend),yh=slice(latbeg,latend),yq=slice(latbeg,latend))
```

```python
month =month.persist()
```

```python
montly_mean = month.groupby("time.month").mean("time").persist()
```
