```python
import ncar_jobqueue
```

```python
cluster = ncar_jobqueue.NCARCluster(
#    cores=9,  # The number of cores you want
    memory='800GB',  # Amount of memory
#    processes=9,  # How many processes
    walltime='01:00:00',  # Amount of wall time
    project="ncgd0011",  #Project id
)

cluster.scale(150)
```

```python
%matplotlib inline
%load_ext autoreload
%load_ext watermark


import dask
import distributed
import matplotlib as mpl
import matplotlib.pyplot as plt

import numpy as np
#import pandas as pd
import xarray as xr
#import xgcm
import cartopy.crs as cr
from cartopy import feature as cf

mpl.rcParams["savefig.dpi"] = 300
mpl.rcParams["savefig.bbox"] = "tight"
mpl.rcParams["figure.dpi"] = 180

xr.set_options(keep_attrs=True, display_style="html")

%watermark -iv
```

```python
client = distributed.Client(cluster)
client
```

```python
files = '/glade/campaign/cgd/oce/people/bachman/ETP_1_20_tides/DAILY/ocean_daily__*.nc'
```

```python
DS = xr.open_mfdataset(
    files,
    compat="override",
    data_vars="minimal",
    coords="minimal",
    parallel=True,
    concat_dim="time",
    combine="nested",
    chunks={"xh": 200, "yh": 400, "z_i": 400},
)[["wo"]]
#[["so", "thetao", "uo", "vo", "wo"]]
```

```python
DS
```

```python
A2 = DS.isel(z_i=slice(35))
A2
```

```python
A2_wo = A2["wo"]
A2_wo
```

```python
A2_wo = A2_wo.sel(xh=np.arange(260,283,0.05), yh=np.arange(0,17,0.05), method='nearest')
```

```python
mes_A2 = A2_wo.groupby('time.month').mean('time').persist()
mes_A2
```

```python
month_vertical_v = mes_A2.to_netcdf('/glade/scratch/pmora/month_w_zi_0-85m.nc')
```

```python
mes_A2.plot.contourf(xh='lon',yh='lat',col='month',col_wrap=2)
```

```python
count = 0
fig=plt.figure(figsize=(13,11))
for month in mes_A2.month:
             count+=1
             ax=fig.add_subplot(4,3,count,projection=cr.PlateCarree())
             mes_A2[mes_A2.month==month].plot(cmap="coolwarm", cbar_kwargs={"label": ""})
             
             ax.coastlines(resolution='10m')
             ax.add_feature(cf.BORDERS)
             ax.set_extent([-100, -77, 0,17])
             
            
#             gl=ax.gridlines()
#             ax.xlabels_botton=True; ax.ylabels_left=True
#            gl.xformatter = LONGITUDE_FORMATTER
#             gl.yformatter = LATITUDE_FORMATTER
```

```python
count = 0
fig=plt.figure(figsize=(11,7))
for mes in mes_A2.month:
             count+=1
             ax=fig.add_subplot(2,2,count,projection=cr.PlateCarree())
             mes_A2[mes_A2.month==mes].plot(vmin=-3,vmax=3, cmap="jet_r", cbar_kwargs={"label": ""} , levels = np.linspace(-3, 3, 7))
             
             ax.coastlines(resolution='10m')
             ax.add_feature(cf.BORDERS)
             ax.set_extent([-88, -83, 8,12.5])
```
