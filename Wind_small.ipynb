```python
path = "/glade/p/cesmdata/cseg/inputdata/ocn/jra55/v1.3_noleap/JRA.v1.3.u_10.TL319.2014.171019.nc"
ds = xr.open_mfdataset(path, parallel=True)  
ds
```

```python
Sel_u = ds['u_10']
U_area = Sel_u.sel(longitude=np.arange(250,290,0.5), latitude=np.arange(5,18,0.5), method='nearest')
U_area
```

```python
Season_U = U_area.groupby('time.season').mean('time')
```

```python
Season_U
```

```python
Season_U.plot.contourf(x='longitude',y='latitude',col='season')
```

```python
count = 0
fig=plt.figure()
for s in Season_U.season:
    count+=1
    ax=fig.add_subplot(2,2,count,projection=cr.PlateCarree())
    Season_U[Season_U.season == s].plot()
    
    ax.coastlines(resolution='110m')
    ax.add_feature(cf.BORDERS)
    ax.set_extent([-90,-85.50,5,18])
```
